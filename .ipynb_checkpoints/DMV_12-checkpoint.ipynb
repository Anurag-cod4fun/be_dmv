```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("customer_shopping_data.csv")
df.head()
```

```python
# To check the count of records grouped by region/branch of the mall
df.groupby("shopping_mall").count()
```

```python
# To check the count of records grouped by the product categories

df.groupby("category").count()
```

```python
# total sales for each category of product
df.groupby("category").sum()
```

```python
#to get the top performing branches

df.sort_values(by = "price", ascending = False)
```

```python

# to get the top selling categories
df = df.sort_values(by='price', ascending=False)

```

```python
# to get total sales for each combination of branch and product_category
df.groupby(["shopping_mall", "category"]).sum()
```

```python
sales_by_category = df.groupby('category')['price'].sum()
sales_by_category
```

```python
# pie chart for sales by category

plt.figure(figsize=(8, 8))
plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f%%')
plt.title('Sales by Category')
plt.show()
```

```python
combined_pivot = df.pivot_table(index="shopping_mall", columns="category", values="price", aggfunc="sum")
```

```python
combined_pivot
```

```python
# grouped bar chart for sales of different categories at different branches

combined_pivot.plot(kind="bar", figsize=(10, 6))
plt.show()
```

```python
# Gender distribution
gender_distribution = df['gender'].value_counts()
plt.bar(gender_distribution.index, gender_distribution.values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution of Customers')
plt.show()
```

```python
# Age distribution
plt.hist(df['age'], bins=10, edgecolor='k')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution of Customers')
plt.show()
```

```python
# Payment gender_distribution
payment_distribution = df['payment_method'].value_counts()
plt.bar(payment_distribution.index, payment_distribution.values)
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.title('Payment Method Distribution')
plt.xticks(rotation=45)
plt.show()
```
